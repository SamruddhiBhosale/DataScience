```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("AvgWaitingTimeJan.csv")

# Sort the data by day
sorted_data = data.sort_values(by='day')

# Extract sorted days and average waiting time
sorted_days = sorted_data['day']
sorted_avg_waiting_time = sorted_data['avg(request_to_pickup)']

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(sorted_days, sorted_avg_waiting_time, color='skyblue')
plt.xlabel('Day')
plt.ylabel('Average Waiting Time')
plt.title('Average Waiting Time by Day')
plt.xticks(sorted_days)
plt.tight_layout()
plt.show()
```
