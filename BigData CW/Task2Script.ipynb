```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = 'monthlytrips.csv'
df = pd.read_csv(file_path)


df['business_month'] = df['business'] + '-' + df['month'].astype(str)


plt.figure(figsize=(10, 6))
plt.bar(df['business_month'], df['trip_count'], color='skyblue')
plt.yscale('log')

for i, trip_count in enumerate(df['trip_count']):
    plt.text(i, trip_count, str(trip_count), ha='center', va='bottom')

plt.xlabel('Business-Month')
plt.ylabel('Trip Counts')
plt.title('Trip Counts by Business and Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("monthly_profit.csv")

df['business_month'] = df['business'] + '-' + df['month'].astype(str)

plt.figure(figsize=(10, 6))
plt.bar(df['business_month'], df['platform_profit'], color='skyblue')


for i, profit in enumerate(df['platform_profit']):
    plt.text(i, profit, f"{profit:.2f}", ha='center', va='bottom')

plt.xlabel('Business-Month')
plt.ylabel('Platform Profits')
plt.title('Platform Profits by Business and Month')
plt.xticks(rotation=45)
plt.tight_layout() 
plt.show()
```
