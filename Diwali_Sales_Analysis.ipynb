# Exploratary Data Analysis

```python
ax = sns.countplot(x = 'Gender' , data = df)

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
gender_amt  = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Gender', y = 'Amount', data = gender_amt)
```

```python
sns.countplot(data = df , x ='Age Group' , hue = 'Gender')
```

```python
df.columns
```

```python
marital_status_grp = df.groupby(['Marital_Status'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False)

sns.barplot(x = 'Marital_Status' , y = 'Orders', data = marital_status_grp)
```

```python
occupation_amountspent = df.groupby('Occupation', as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc = {'figure.figsize' : (20,5)})
sns.barplot(x = 'Occupation' , y = 'Amount' , data = occupation_amountspent )
```

```python
occupation_orders = df.groupby('Occupation', as_index= False)['Orders'].sum().sort_values(by = 'Orders', ascending = False)
sns.set(rc = {'figure.figsize' : (20,5)})

sns.barplot(x = 'Occupation' , y='Orders', data = occupation_orders)
```
